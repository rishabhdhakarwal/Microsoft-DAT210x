```python
import pandas as pd
```

```python
df=pd.read_csv("/home/rishabh/Desktop/Data Science/DAT210x-master/Module2/Datasets/census.data", names=['education', 'age', 'capital-gain', 'race', 'capital-loss', 'hours-per-week', 'sex', 'classification'], na_values='?')
```

```python
df.education.unique()
```

```python
ordered_education = ['5th', '6th', '7th', '8th','7th-8th', '9th', '10th', '11th', '12th', 'HS-grad', 'Some-college' , 'Bachelors','Masters','Doctorate' ]
df.education = df.education.astype("category", ordered=True, categories=ordered_education).cat.codes
```

```python
classification_ordered = [
    '<=50K', '>50K']
df.classification = df.classification.astype('category',
                                   ordered=True,
                                   categories=classification_ordered).cat.codes
```
